```python
from sionna.rt import load_scene, Transmitter, Receiver, PlanarArray, PathSolver
import os

scene = load_scene(os.path.expanduser(
    "~/ns-allinone-3.40/ns-3.40/contrib/sionna/model/ns3sionna/models/warehouse/warehouse.xml"),
    merge_shapes=False)
scene.frequency = 5.18e9
scene.bandwidth = 20e6

scene.add(Transmitter("drone_0", position=[-3.0, 0.0, 4.0]))
scene.add(Receiver("drone_1", position=[0.0, 0.0, 5.0]))
scene.add(Receiver("drone_2", position=[3.0, 0.0, 6.0]))

scene.tx_array = PlanarArray(num_rows=1, num_cols=1,
                              vertical_spacing=0.5, horizontal_spacing=0.5,
                              pattern="dipole", polarization="V")
scene.rx_array = PlanarArray(num_rows=1, num_cols=1,
                              vertical_spacing=0.5, horizontal_spacing=0.5,
                              pattern="dipole", polarization="V")

paths = PathSolver()(scene=scene, max_depth=3, samples_per_src=10**6,
                     los=True, specular_reflection=True, refraction=True)

scene.preview(paths=paths, show_devices=True, clip_at=5.9)
```
